```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import time
from datetime import datetime
from datetime import date
```

```python
df1=pd.read_csv('../datasets/chelseapalacetweets.csv')
df2=pd.read_csv('../datasets/chelseapalacetweets2.csv')
df3=pd.read_csv('../datasets/chelseapalacetweets3.csv')
df4=pd.read_csv('../datasets/chelseapalacetweets4.csv')
df5=pd.read_csv('../datasets/chelseapalacetweets5.csv')
df6=pd.read_csv('../datasets/chelseapalacetweets6.csv')
df7=pd.read_csv('../datasets/chelseapalacetweets7.csv')
```
